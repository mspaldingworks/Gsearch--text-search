# Parsing text file with regex and pandas

## Import Libraries and Load Data
Common Python libraries are pandas and numpy for data manipulation, matplotlib and seaborn for data visualization, regular expressions to extract key patterns from TXT file, and sys for file system functions.

If you need to install any of these packages for the first time, please follow the steps in this tutorial to pip install as needed: https://realpython.com/what-is-pip/

This program uses a local file from my computer, which details my communications with collaborators over the course of almost two years as we worked together to create my thesis exhibition.



```python
import sys
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


with open('/Users/madelynspalding/Gsearchdata.txt', 'r') as f:
    
    f_contents = f.read()
    
    print(f_contents)  
```

```python
email_chains = re.compile(r'DS\s1\.\d{1,3}') #pull out numbered datasources under 'DS' in text file

matches = email_chains.findall(f_contents)

for match in matches:
    
    print(match)
```

```python
dates_pattern1 = re.compile(r'(\d{1,2}/\d{1,2}/\d{2})- (\d{1,2}/\d{1,2}/\d{2})', re.MULTILINE)

matches = dates_pattern1.findall(f_contents)

for match in matches: 
   
    print(match)

#for match in matches:
 #   print(pattern_combined)
```

```python

line = f_contents

match = re.findall(r'[\w\.-]+@[\w\.-]+', line)

match
```
